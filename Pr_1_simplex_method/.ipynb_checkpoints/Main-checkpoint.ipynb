```python
def count_target_function(x,y):
    target_function = x**2 - x*y + 3*y**2-x
    return round(target_function, 3)
```

```python
def condition_for_the_end_of_search(mass, mass_masximum, e):
    center_sim = []
    for j in range(n):
        coordinate = 0
        for i in range(len(mass)):
            if i not in mass_masximum:
                coordinate += mass[i][j]
        coordinate = coordinate / 3
        center_sim.append(coordinate)
    target = count_target_function(center_sim[0], center_sim[1])
    count_f = 0
    count_true = 0
    for i in range(len(mass)):
        if i not in mass_masximum:
            funck = mass[i][-1] - target
            count_f += 1
            if funck < e:
                count_true += 1
    return count_f == count_true
```

```python
print(condition_for_the_end_of_search(mass, mass_masximum, e))
```
